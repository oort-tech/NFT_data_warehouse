- one_day_volume: The total trading volume for the NFT in the last 24 hours.  
- one_day_change: The percentage change in price of the NFT in the last 24 hours.
- one_day_sales: The total number of sales of the NFT in the last 24 hours.
- one_day_average_price: The average price of the NFT in the last 24 hours.
- seven_day_volume: The total trading volume for the NFT in the last 7 days.
- seven_day_change: The percentage change in price of the NFT in the last 7 days.
- seven_day_sales: The total number of sales of the NFT in the last 7 days.
- seven_day_average_price: The average price of the NFT in the last 7 days.
- thirty_day_volume: The total trading volume for the NFT in the last 30 days.
- thirty_day_change: The percentage change in price of the NFT in the last 30 days.
- thirty_day_sales: The total number of sales of the NFT in the last 30 days.
- thirty_day_average_price: The average price of the NFT in the last 30 days.
- total_volume: The total trading volume for the NFT since it was minted.
- total_sales: The total number of sales of the NFT since it was minted.
- total_supply: The maximum number of NFTs that can be minted in this collection.
- total_minted: The total number of NFTs that have been minted in this collection.
- num_owners: The number of unique addresses that own this NFT.
- average_price: The average price at which this NFT has been sold.
- market_cap: The total value of all NFTs in this collection, calculated as total_supply * average_price.
- floor_price: The lowest price at which this NFT is currently being sold.
- floor_price_historic_one_day: The lowest price at which this NFT was sold in the last 24 hours.
- floor_price_historic_seven_day: The lowest price at which this NFT was sold in the last 7 days.
- floor_price_historic_thirty_day: The lowest price at which this NFT was sold in the last 30 days.
- updated_date: The date and time at which these statistics were last updated.

```python
# Retrieving Contract sales stat and save as csv
import requests
from datetime import datetime
import os
import pandas as pd
import json


chain_dic = ['ethereum','polygon']
contract_address_Eth = {'Decentraland':'0xf87e31492faf9a91b02ee0deaad50d51d56d5d4d',
                   'Bored Ape Yacht Club':'0xBC4CA0EdA7647A8aB7C2061c2E118A18a936f13D'}
contract_address_Pol = {
    'The Martians: Metaverse':'0x09aA1A14e572e6fC05F64e541699D3E0361C7F2f',
    'Chumbi Valley Official':'0x5492Ef6aEebA1A3896357359eF039a8B11621b45'
}


for chain in chain_dic:
    url = f"https://api.nftport.xyz/v0/transactions/stats/0xbc4ca0eda7647a8ab7c2061c2e118a18a936f13d?chain={chain}"

    headers = {
        "accept": "application/json",
        "Authorization": "bdcedc42-0148-44cf-8912-7119a36dfc2d"
    }

    response = requests.get(url, headers=headers)
    
    # Convert the JSON string to a dictionary
    response_dict = json.loads(response.text)
    
    # Extract the statistics from the dictionary
    statistics_dict = response_dict['statistics']

    # Convert the statistics dictionary to a DataFrame
    df_statistics = pd.DataFrame.from_dict(statistics_dict, orient='index').T
    
    today = datetime.today().strftime('%Y_%m_%d')
    
    filename = f'sales_static_{chain}_{today}.csv'
    
    if os.path.exists(filename):
        print(f'{filename} already exists. Removing...')
        os.remove(filename)
    
    df_statistics.to_csv(filename, index=False)
```
