###Breaking rarity into columns

```python
# Define a function to extract values from the 'rarity' column
def extract_rarity_values(row):
    row = row.strip("{}") # Remove the outer curly braces
    values = row.split(", ") # Split the string into a list of key-value pairs
    values_dict = {}
    for value in values:
        key, val = value.split(": ") # Split each key-value pair into separate key and value strings
        values_dict[key.strip("'")] = val # Remove the single quotes around the keys and assign the value to the dictionary
    return pd.Series(values_dict)

# Apply the function to the 'rarity' column
df_Decentraland_dropped[['strategy', 'score', 'rank', 'collection_size', 'updated_date']] = df_Decentraland_dropped['rarity'].apply(extract_rarity_values)

df_Decentraland_dropped.drop('rarity',axis = 1)
```

#df_ape pre-processing

```python
to_drop_ape = ['chain','contract_address','metadata_url','animation_url','cached_file_url','cached_animation_url','owner']
df_ape_drop = df_Ape.drop(to_drop_ape,axis = 1)
```

```python
df_ape_drop['rarity'][0]
```

```python
import json
df = df_ape_drop
# Replace all single quotes with double quotes in the rarity column
df['rarity'] = df['rarity'].apply(lambda x: x.replace("'", '"'))
# Load the rarity column as a JSON object
df['rarity'] = df['rarity'].apply(lambda x: json.loads(x))

# Flatten the rarity column using json_normalize()
rarity_df = pd.json_normalize(df['rarity'])

# Concatenate the flattened rarity dataframe with the original dataframe
df = pd.concat([df, rarity_df], axis=1)

# Drop the original rarity column
df = df.drop('rarity', axis=1)
```

```python
def extract_attributes(metadata):
    metadata_dict = {}
    if metadata is None:
        return {
                'metadata_Earring': '',
                'metadata_Background': '',
                'metadata_Fur': '',
                'metadata_Clothes': '',
                'metadata_Mouth': '',
                'metadata_Eyes': '',
                'metadata_Hat': ''}
    for trait in metadata['attributes']:
        metadata_dict[f"metadata_{trait['trait_type']}"] = trait['value']
    return metadata_dict
```

```python
def parse_metadata(x):
    try:
        return json.loads(x.replace("'", "\""))
    except json.JSONDecodeError as e:
        print(f"Error parsing metadata: {x}")
        return None
```

```python
# Apply the extract_attributes() function to each row using apply()
df_attributes = df['metadata'].apply(lambda x: extract_attributes(parse_metadata(x)))

df_attributes = pd.json_normalize(df_attributes)
df = pd.concat([df, df_attributes], axis=1)

# df = df.drop(['metadata', 'image', 'attributes'], axis=1)


# Drop the original metadata column
# df = df.drop('metadata', axis=1)
```
